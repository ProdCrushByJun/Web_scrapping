```python
import requests
import time
from bs4 import BeautifulSoup as BS
from time import sleep
import pandas as pd
```

```python
seasons = []
for season in range(1,5): # 5 is seasons count
    season = f'https://www.imdb.com/title/tt4574334/episodes?season={season}'
    seasons.append(season)
```

```python
links = []
titles = []
stars = []
descriptions = [] 
episode_numbers = []
for ep in range(len(seasons)):
    response = requests.get(seasons[ep],'xmlx')
    soup = BS(response.text,'lxml')
    data = soup.find_all('div',class_='info')
    for episode in data:
        link = 'https://www.imdb.com' + episode.find("a").get('href')
        title = episode.find("a").get('title')
        star = episode.find('span',class_="ipl-rating-star__rating").text.strip()
        description = episode.find('div',class_="item_description").text.strip()
        season_number = seasons[ep][-1]
        links.append(link)
        titles.append(title)
        stars.append(star)
        descriptions.append(description)
        season_numbers.append(season_number)
```

```python
durations = []
episode_numbers=[]
for k in range(len(links)):
    response = requests.get(links[k],'xmlx')
    soup = BS(response.text,'lxml')
    episode_number = soup.find('div',class_="sc-2a366625-1 cHCfvp").text
    dur_data = soup.find_all('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-none ipc-metadata-list--compact ipc-metadata-list--base")
    for item in dur_data:
        duration = item.find('div',class_="ipc-metadata-list-item__content-container").text
        durations.append(duration)
        episode_numbers.append(episode_number)
```

```python
df = pd.DataFrame(list(zip(season_numbers,episode_numbers,links, titles,stars,descriptions,durations)),
                  columns = ['Season','Ep.num','Url', 'Name', 'Rating', 'Desc','Duration'])
df
```

```python
df.to_csv('StrangerThings.csv', index=False)
```
